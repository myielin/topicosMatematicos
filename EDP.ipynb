```python
import math as m
import pandas as pd
import plotly.express as px
```

```python
L = 20
Ti = 0
Tf = 0
N = 1000
```

```python
def v(x):
  return -(x**2)/10 +2*x
```

```python
def w(x,t):
  w = 0
  for n in range(1, N):
    e = (-(n**2*m.pi**2/L**2))*t
    w += (160/(n**3*m.pi**3))* ((-1)**n - 1)*(m.e**(e))*m.sin((n*m.pi*x)/L)
  return w
```

```python
def u(x,t):
  return v(x) + w(x,t)
```

```python
yv = [v(x) for x in range(21)]
px.line(x = range(21), y = yv, title = "Distribluição de calor estacionária v(x)")
```

```python
dfw = pd.DataFrame()
for i in range(0, 55, 5):
    dfw[f"t = {i}"] = [w(x, i) for x in range(21)]
```

```python
yw = [f"t = {i}" for i in range(0, 55, 5)]
px.line(dfw, x = range(21), y = yw, title = "distribuição de calor transiente w(x,t)")
```

```python
y = [u(x, 0.0) for x in range(21)]
px.scatter(x = range(21), y = y, title = "instante 0")
```
