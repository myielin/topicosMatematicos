```python
from math import e
import pandas as pd
import matplotlib.pyplot as plt
```

#### edo: y' = ay

y = c * e^at<br>
y' = [c * e^at] * at

```python
y0 = [-1, -0.1, 0, 0.1, 1]        # condições iniciais
a = [-1, -2, 0, 1, 2]             # Coeficiente
t = [i*0.1 for i in range(16)]    # intervalo
```

t = [0:16] <br>
h = 0.1

```python
def y(y0, a):
  return [y0 * e**(a*(h)) for h in t]
```

```python
def euler(y, a):
  return [y[i] + 0.1*(a*y[i]) for i in range(16)]
```

```python
el = euler(y(y0[4], a[1]), a[1])
el
```

```python
y(y0[4], a[1])
```

## resultado analítico
cada variável corresponde a uma lista contendo todos os possíveis diferentes valores iniciais (y0) e os valores fixos de a

```python
a0 = [y(y0[i], a[0]) for i in range(5)]   # a fixo igual a -1 e todas possíveis condições y0
a1 = [y(y0[i], a[1]) for i in range(5)]
a2 = [y(y0[i], a[2]) for i in range(5)]
a3 = [y(y0[i], a[3]) for i in range(5)]
a4 = [y(y0[i], a[4]) for i in range(5)]
```

resultado de euler

```python
ea0 = [euler(y(y0[i], a[0]), a[0]) for i in range(5)]
ea1 = [euler(y(y0[i], a[1]), a[1]) for i in range(5)]
ea2 = [euler(y(y0[i], a[2]), a[2]) for i in range(5)]
ea3 = [euler(y(y0[i], a[3]), a[3]) for i in range(5)]
ea4 = [euler(y(y0[i], a[4]), a[4]) for i in range(5)]
```

```python
for i in range(5):
  plt.plot(t, a0[i])
  plt.plot(t, ea0[i])
```

```python
for i in range(5):
  plt.plot(t, a4[i])
  plt.plot(t, ea4[i])
```

Para que y(0) = 1 e y'(0) = 0 com a = 1:<br>
 y = e^0 * c   ------------->  c = 1 <br>
 y' = e^0 * c   -------------> c = 1

```python
px.line(x = t, y = [y(1, 1), euler(y(1, 1), 1)])
```
