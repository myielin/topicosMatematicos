M(x, N) = Σ dₙ * sen(nx)

```python
x = [i/10 for i in range(-10,10)]
def m_b(func, N):
  y = []

  for x in range(-10, 10):
    r = 0
    for i in range(N):
      r += func(i, h=0.00001) * m.sin(i * x/10)
    y.append(r)

  return y
```

## Função: ∫ sen(nx) dx
[0:1]

```python
def n1(n, h=0.0001):
  pr = 0               # resultado da função com peso
  i = 0
  while i <= 1:
    pr += 2 * m.sin(n*i)
    i += h

  return (h/2) * pr
```

```python
print_result(n1, 1, nw=0)
```

```python
print_result(n1, 5, nw=1)
```

```python
print_result(n1, 10, nw=2)
```

## função ∫x*sen(nx)dx

```python
def n2(n, h=0.0001):
  pr = 0               # resultado da função com peso
  i = 0
  while i <= 1:
    pr += 2 * (i * m.sin(n*i))
    i += h

  return (h/2) * pr
```

```python
print_result(n2, 1, nw=4)
```

```python
print_result(n2, 5, nw=5)
```

```python
print_result(n2, 10, nw=6)
```

```python
print_result(n2, 100, nw=7)
```

## função ∫x² * sen(nx)dx

```python
def n3(n, h=0.0001):
  pr = 0               # resultado da função com peso
  i = 0
  while i <= 1:
    pr += 2 * (i*i * m.sin(n*i))
    i += h

  return (h/2) * pr
```

```python
print_result(n3, 1, nw=8)
```

```python
print_result(n3, 5, nw=9)
```

```python
print_result(n3, 10, nw=10)
```

```python
print_result(n3, 100, nw=11)
```
